```python
def eda_plot(column, topn=10):
    plt.figure(figsize=(4, 2))
    df[column].value_counts().nlargest(topn).plot(kind='bar')
    plt.title(f"'{column}' column value counts")
    plt.ylabel("Counts")
    plt.xticks(rotation = 45, ha='right')
    plt.show()

columns = ['title', 'company_name', 'location', 'via', 'schedule_type', 'work_from_home']

for column in columns:
    eda_plot(column)
```

```python
missing_value = pd.Series(df.isnull().sum())
missing_percentage = pd.Series(df.isnull().mean()*100)
```

```python
missing = pd.concat([missing_value,missing_percentage] , axis = 1)
missing.columns = ['missing_values','missing_percentage']
missing.sort_values('missing_percentage' ,ascending = False)
```

# Data Cleaning
---

```python
# Create a list of columns with high missing values
miss = missing.loc[missing['missing_percentage'] >= 30 , : ].reset_index()
miss = list(miss['index'])
miss
```

```python
# Drop unnecssary columns and columns with too many missing values however some columns.
# However , some columns with many missing values such as work_from_home and salary standardized are worth keeping and will be used in the analysis
df = df.drop(columns = ['thumbnail',
  'title',
 'salary',
 'commute_time',
 'salary_pay',
 'salary_rate',
 'salary_avg',
 'salary_min',
 'salary_max',
 'salary_hourly',
 'salary_yearly',
 'index', 
  'via' ,
  'description' ,
  'extensions' , 
  'job_id',
  'posted_at' , 
  'Unnamed: 0' ])
```

```python
# Clean location and description_tokens columns.

df[['location1' , 'location2']] = df.location.str.split(',' , expand = True)
df = df.drop(columns = ['location' , 'location2'])
```

```python
df['description_tokens'] = df.description_tokens.str.replace("'","").str.replace('[' , "").str.replace(']' , '').str.replace(' ','')
program = df['description_tokens'].str.get_dummies(',')
program = program.sum().sort_values(ascending = False)
program = pd.DataFrame(program)
program.columns = ['Count']
program = program.head(20)
program
```

# Data visualization
---

```python
# Plot top skills / softwares in data analyst
fig = plt.figure(figsize = (10,5))
ax = sns.barplot(data = program , y = program.index , x = program['Count'])
ax.set_xlabel('Softwares / Programming Languages')
ax.set_title('Softwares or Programming Languages which Data Analyst used in their daily lives')
ax.set_xticklabels(labels = [] ,rotation = 45)
ax.tick_params(axis="y", labelsize=11);
sns.despine(right=True, top=True, bottom=True)

for i, v in enumerate(program[:20].values):
    ax.text(v, i, str(v), color="white", ha="right", va="center_baseline")

```

```python
# Plot annual income distributiion

fig = plt.figure(figsize = (10,8))
plt.suptitle('Range of Salary')

df_sched = df[df['schedule_type'] != 'Internship']
ax3 = fig.add_subplot(2,1,2)
ax3 = sns.boxplot(df_sched  , x = 'salary_standardized' , y = 'schedule_type')
ax3.set_ylabel('Schedule Type')
ax3.set_xlabel('Salary($) per year')
ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)
ax3.spines['left'].set_visible(False)
ax3.spines['bottom'].set_visible(False)
ax3.set_title('Range of Salary by  Schedule Type')

ax1 = fig.add_subplot(2,2,1)
ax1 = sns.histplot(df , x=  'salary_standardized' )
ax1.set_title('Salary Distribution')
ax1.set_xlabel('')

ax2 = fig.add_subplot(2,2,2)
ax2 = sns.boxplot(df,x = df.salary_standardized)
ax2.set_xlabel('Salary($) per year')
ax2.set_title('Range of Salary')
plt.tight_layout()

plt.savefig('2.SalaryRange.png')
```
